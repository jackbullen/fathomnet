```python
import numpy as np
import PIL.Image

from io import BytesIO

img = PIL.Image.open(BytesIO(req.content))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
def plot_image(url):
    req = requests.get(url.url)
    img = PIL.Image.open(BytesIO(req.content))
    plt.figure()
    plt.imshow(img)
    for bbox in url.boundingBoxes:
        plt.gca().add_patch(plt.Rectangle((bbox.x, bbox.y), bbox.width, bbox.height, fill=False, edgecolor='r', linewidth=2))


for url in images.find_by_concept('Nanomia')[0:10]:
    plot_image(url)
```
